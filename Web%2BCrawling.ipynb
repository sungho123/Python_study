````python
def image_url_crawl(articles):
    image = []
    for article in articles:
        dls = article.findAll('dl')
        for dl in dls:
            try:
                img_url = dl.img['src']
                image.append(img_url)
            except:
                image.append("No image")
    return(pd.DataFrame(image, columns=['image_url']))
````

````python
#기사 제목  & url
def title_url_crawl(articles):
    title = []
    for article in articles:
        dls = article.findAll('dl')
        for dl in dls:
            dt = dl.find('dt', {'class':''})
            #title
            title_text = re.sub('\n|\r|\t','',dt.text).strip()
            #url
            url = dt.a['href']

            title.append([title_text, url])
    return(pd.DataFrame(title, columns=['title','url']))
````

````python
def rest_crawl(articles):
    output = []
    for article in articles:
        dls = article.findAll('dl')
        for dl in dls:
            dd_temp = dl.dd
            #기사내용
            try:
                body = re.sub('\r|\n|\t','', dd_temp.text).split('…')[0]
            except:
                body = "None"
            #언론사
            try:
                nsp = dd_temp.find("span", {"class":"writing"}).text
            except:
                nsp = "None"
            #날짜
            try:
                date = dd_temp.find("span", {"class":"date"}).text
            except:
                date = "None"
            
            output.append([body, nsp, date])
    return(pd.DataFrame(output, columns=['body','nsp','date']))
````

````python
rest_crawl(articles) #마지막은 에러남..
````

````python
def get_navernews(url):
    html = urlopen(url)
    soup = BeautifulSoup(html, "html.parser")
    
    articles = soup.findAll("li")
    
    title_url = title_url_crawl(articles)
    con_wrt_date = rest_crawl(articles)
    img_url = image_url_crawl(articles)
    
    total = pd.concat([title_url, con_wrt_date, img_url], 1) #cbind와 동일 (0이면 rbind)
    return total
````

````python
def get_naver_realasset(area_code):
    url = 'http://land.naver.com/article/articleList.nhn?' \
        + 'rletTypeCd=A01&tradeTypeCd=A1&hscpTypeCd=A01%3AA03%3AA04' \
        + '&cortarNo=' + area_code

    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')

    table = soup.find('table')
    trs = table.tbody.find_all('tr')
    
    value_list = []

    # 거래, 종류, 확인일자, 매물명, 매물명, 면적(㎡), 층, 매물가(만원), 연락처
    for tr in trs[::2]:
        tds = tr.find_all('td')
        cols = [re.sub('\n|\t|\r','',i.text) for i in tds]


        if '_thumb_image' not in tds[3]['class']: # 현장확인 날짜와 이미지가 없는 행
            cols.insert(3, '') 

        거래 = cols[0]
        종류 = cols[1]
        확인일자 = datetime.strptime(cols[2], '%y.%m.%d.')
        현장확인 = cols[3]
        매물명 = cols[4]
        면적 = cols[5]
        공급면적 = re.findall('공급면적(.*?)㎡', 면적)[0]
        전용면적 = re.findall('전용면적(.*?)㎡', 면적)[0]
        공급면적 = float(공급면적)
        전용면적 = float(전용면적)
        층 = cols[6]
        
        if cols[7].find('호가일뿐') >= 0:
            매물가 = "호가"
        else:
            매물가 = int(cols[7].replace(',', ''))
            
        연락처 = cols[8]
        
        value_list.append([거래, 종류, 확인일자, 현장확인, 매물명, 공급면적, 전용면적, 층, 매물가, 연락처])
        
    cols = ['거래', '종류', '확인일자', '현장확인', '매물명', '공급면적', '전용면적', '층', '매물가', '연락처']
    df = pd.DataFrame(value_list, columns=cols)
    return df
````

````python
#여러 페이지의 값을 불러오도록
def get_naver_realasset(area_code, page_num):
    #페이지만 늘어나는데 뒤에 속성이 많이 붙음..
    url = 'http://land.naver.com/article/articleList.nhn?' \
        + 'rletTypeCd=A01&tradeTypeCd=A1&hscpTypeCd=A01%3AA03%3AA04' \
        + '&cortarNo=' + area_code \
        + '&articleOrderCode=&siteOrderCode=&cpId=&mapX=&mapY=&mapLevel=&minPrc=&maxPrc=&minWrrnt=&maxWrrnt=&minLease=&maxLease=&minSpc=&maxSpc=&subDist=&mviDate=&hsehCnt=&rltrId=&mnex=&mHscpNo=&mPtpRange=&mnexOrder=&location=2538&ptpNo=&bssYm=&schlCd=&cmplYn=&page={}#_content_list_target'.format(page_num) 

    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')

    table = soup.find('table')
    trs = table.tbody.find_all('tr')
    if '등록된 매물이 없습니다' in trs[0].text:
        return pd.DataFrame()
    
    value_list = []

    # 거래, 종류, 확인일자, 매물명, 면적(㎡), 층, 매물가(만원), 연락처
    for tr in trs[::2]:
        tds = tr.find_all('td')
        cols = [' '.join(td.text.strip().split()) for td in tds]

        if '_thumb_image' not in tds[3]['class']: # 현장확인 날짜와 이미지가 없는 행
            cols.insert(3, '')

        # print(cols)
        거래 = cols[0]
        종류 = cols[1]
        확인일자 = datetime.strptime(cols[2], '%y.%m.%d.')
        현장확인 = cols[3]
        매물명 = cols[4]
        면적 = cols[5]
        공급면적 = re.findall('공급면적(.*?)㎡', 면적)[0]
        전용면적 = re.findall('전용면적(.*?)㎡', 면적)[0]
        공급면적 = float(공급면적)
        전용면적 = float(전용면적)
        층 = cols[6]
        
        if cols[7].find('호가일뿐') >= 0:
            매물가 = "호가"
        else:
            매물가 = int(cols[7].replace(',', ''))
            
        연락처 = cols[8]
        
        value_list.append([거래, 종류, 확인일자, 현장확인, 매물명, 공급면적, 전용면적, 층, 매물가, 연락처])
        
    cols = ['거래', '종류', '확인일자', '현장확인', '매물명', '공급면적', '전용면적', '층', '매물가', '연락처']
    df = pd.DataFrame(value_list, columns=cols)
    return df
````
