## 02. glob

glob 모듈은 윈도우의 dir명령어나 리눅스의 1s명령어와 유사한 기능 제공


```python
#glob.glob(path)-경로에 대응되는 모든 파일 및 디렉터리의 리스트 반환
## *와 ?도 사용가능, []를 사용한 문자열 비교도 가능

import glob
glob.glob('canon?.*')
glob.glob('gayoung[0-9].*')
```

```python
#경로를 주는 방식에 따라 절대 경로로 결과가 나오게 할 수도 있음

glob.glob('*.exe')
glob.glob(abspath('.')+'\\*.exe')
```

```python
#glob.iglob(path)-glob과 동일한 동작을 수행하지만 리스트로 결과를 반환하지 않고 이터레이터를 반환

glob.iglob('*')


# In[ ]:

for i in glob.iglob('*'):
    print(i)
```

# 03. tree 예제

tree는 하위 디렉터리 구조를 보여주는 툴  
os.path와 glob의 일부 기능을 이용해 tree와 비슷한 기능을 수행하는 예제를 작성할 수 있다  
    
예제: traverse 함수(위치를 지정해주면 그 안의 모든 디렉토리와 파일 출력)를 만들고, 함수 내부에서는 파일인 경우에는 파일 이름을  
디렉터리인 경우에는 디렉터리명을 출력한 후 그 안의 내용을 다시 처리

```python
import glob, os.path
ndir=nfile=0

def traverse(dir,depth):
    global ndir,nfile
    for obj in glob.glob(dir + '/*'):
        if depth==0:
            prefix='|--'
        else:
            prefix='|' + ' '*depth + '|--'
        if os.path.isdir(obj): #디렉터리인 경우
            ndir += 1
            print(prefix + os.path.basename(obj))
            traverse(obj, depth+1)
        elif os.path.isfile(obj): #파일인 경우
            nfile += 1
            print(prefix + os.path.basename(obj))
        else: #그 밖의 경우 처리
            print(prefix + 'unknown objec :', obj)
            
if __name__=='__main__':
    traverse('.',0) #처음에는 현재 디렉토리로 지정
    print('\n', ndir, 'directories', 'nfile', 'files') #전체 합계 출력
```
